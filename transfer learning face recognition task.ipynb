```python
from keras.applications import MobileNet
from glob import glob
from keras.models import Model
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import numpy as np 
import matplotlib.pyplot as plt
from keras.layers import Input ,Lambda, Dense, Flatten
```

```python
model=MobileNet(weights='imagenet',include_top=False,input_shape=(224,224,3))
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
trainpath='dogbreed/train'
testpath='dogbreed/test'
```

```python
folders=glob('monkey_breed/monkey_breed/train/*')
```

```python
topmodel=model.output
```

```python
topmodel=Flatten()(model.output)
```

```python
topmodel=Dense(1024,activation='relu')(topmodel)
```

```python
topmodel=Dense(1000,activation='relu')(topmodel)
```

```python
topmodel=Dense(len(folders),activation='softmax')(topmodel)
```

```python
newmodel=Model(inputs=model.input,outputs=topmodel)
```

```python
newmodel.output
```

```python
newmodel.summary()
```

```python
newmodel.compile(loss='categorical_crossentropy',
                optimizer='adam',
                metrics=['accuracy']
                )
```

```python
newmodel.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
training_set = train_datagen.flow_from_directory(
        'monkey_breed/monkey_breed/train/',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical')
test_set = test_datagen.flow_from_directory(
        'monkey_breed/monkey_breed/validation/',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical')
r = newmodel.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=5,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```
